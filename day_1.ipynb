Question 1

Use the strikes column to find where there are 2 strikes and make a new column indicating whether the pitch was thrown in a 2-strike situation. Name this is_2_strikes

Question 2

Use the release_speed column to find when a pitch is thrown greater than 94 MPH (the column is in MPH already). Name this is_high_velo

```python
df['is_high_velo'] = 0
df.loc[df['release_speed'] > 94, 'is_high_velo' ] = 1
```

Question 3

Use the description column AND the attack_zone column to indicate when a pitch is thrown for a called or swinging strike (not a foul) in the "heart" attack zone (hint: this requires two conditional statements! maybe google or ask for help in the discord!)

```python
onli_strike = ['called_strike', 'swinging_strike']
```

```python
df['is_heart_zone'] = 0
df.loc[(df['description'].isin(onli_strike)) & (df['attack_zone'] == 'heart'), 'is_heart_zone'] = 1
```
